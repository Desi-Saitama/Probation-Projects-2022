--As month column is saved as object we will change it to date and time format hence grouping will be easy.

```python
df["month"] = pd.to_datetime(df["month"])
print(df)
print (df.dtypes)
```

--We will now group data year wise and will try to plot bar graph between all values with respect to each year.

```python
dt = df.groupby(pd.Grouper(key='month', freq='1Y'))['totals'].sum()
print(dt)

#find growth rate for each year with respect to year prior to it.
dt.pct_change()
```

Growth Rate Shows there has been signifcant increase in permits for election year in USA 
EX.
2020: 40%
2016: 19%
2012: 19%
2008: 13%


EXTREME INCREASE IN 1999 IS BECAUSE  U.S. Senate passes a bill requiring trigger locks on all newly manufactured handguns and extending waiting period and background check requirements to sales of firearms at gun shows. 


```python
plt.figure(figsize=(18,10))
dt.pct_change().plot(kind='bar')
```
